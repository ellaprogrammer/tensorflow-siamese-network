```python
def grid_add_bars(img):
    im = plt.imread(img)
    im.setflags(write=1)
    im.flags
    im2 = im
    y = range(0,im2.shape[1])
    im2[:, np.sin(y)<0,:] = 0
    im2 = im
    x = range(0, im2.shape[0])
    y = range(0,im2.shape[1])
    im2[np.sin(x)<0,:,:] = 0
    
    return im2
```

```python
def vertical_add_bars(img):
    im = plt.imread(img)
    im.setflags(write=1)
    im.flags
    im2 = im
    y = range(0,im2.shape[1])
    im2[:, np.sin(y)<0,:] = 0
    
    return im2
```

```python
def diagnol_add_bars(img):
    im = plt.imread(img)
    im_mask = im.copy()
    x = range(0, im_mask.shape[0])
    y = range(0,im_mask.shape[1])
    for i in x:
        for j in y:
            if np.sin((i+j)/10) < -.7:
                im_mask[i,j,:] = 0
            if np.sin((i-j)/10) < -.7:
                im_mask[i,j,:] = 0
    
    return im_mask
```

```python
from PIL import Image 
from matplotlib import pyplot as plt
import numpy as np
import glob
   
 # Relative Path 
imglist = glob.glob("../Desktop/monkeyface/*.jpg")
for img in imglist:
    final = Image.open(img)
    original_name = final.filename
    original_name = (original_name[0:(len(original_name)-4)])
    Image.fromarray( vertical_add_bars(img) ).save(original_name + "_vertical_occluded.jpg") 
    Image.fromarray( grid_add_bars(img) ).save(original_name + "_grid_occluded.jpg") 
    Image.fromarray( diagnol_add_bars(img) ).save(original_name + "_diagnol_occluded.jpg") 
```
